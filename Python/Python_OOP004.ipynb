```python
# Fizz Buzz
for num in range(1,101):
    if num % 5 == 0 and num % 3 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

```python
# Fibonacci Generator
def fib(num):
    a,b = 0,1
    for i in range(0,num):
        yield "{}:{}".format(i+1,a)
        a,b = b, a+b
        
        
for item in fib(10):
    print(item)
```

```python
# Refresh Python OOP concepts : Inheritance + Polymorphism + Encapsulation + Abstraction

class Person(object):
    def __init__(self,name):
        self.name = name
    
    def reveal_identity(self):
        print("My name is {}".format(self.name))

class SuperHero(Person):
    def __init__(self,name,hero_name):
        super(SuperHero,self).__init__(name)
        self.hero_name = hero_name
        
    def reveal_identity(self):
        super(SuperHero,self).reveal_identity()
        print("...And I am {}".format(self.hero_name))

Yang = Person("Yang")
Yang.reveal_identity()

Wade = SuperHero("Wade Wilson","Deadpool")
Wade.reveal_identity()
```

```python
# Create a number square function than implemente it using Python Generator and then implement it using List comprehension 
# Implement it using function 
def square_numbers(nums):
    result = []
    for i in nums:
        result.append(i*i)
    return result

my_nums = square_numbers([1,2,3,4,5])
print(my_nums)

# Implement it using Python Generator 
def square_numbers(nums):
    for i in nums:
        yield (i*i)
        
my_nums= square_numbers([1,2,3,4,5])
for num in my_nums:
    print(num)

# Implement it using Python List Comprehension
my_nums = (x*x for x in [1,2,3,4,5])
print(list(my_nums))                 # if a generator has lots of values that in the list it will lose the performance as them are put in memeroy
```

```python
# create a function compare the exection speed for list and generator
import memory_profiler as mem_profile
import random
import time

names = ['Lei','John','Corey','Adam','Steve','Rick','Thomas','Yang']
majors = ['Math','Engineering', 'Arts','Arts','Computer','History','Math','Math']

print('Memory(Before): {}MB'.format(mem_profile.memory_usage_resource()))

def people_list(num_people):
    result = []
    for i in range(num_people):
        person = {
             'id': i,
             'name':random.choice(names),
             'major':random.choice(majors)
        }
        result.append(person)
    return result

def people_generator(num_people):
    for i in range(num_people):
        person = {
            'id': i,
            'name': random.choice(names),
            'major':random.choice(majors)
        }
        yield person
        
t1 = time.clock()
people = people_list(1000000)
t2 = time.clock()

t1 = time.clock()
people = people_generator(1000000)
t2 = time.clock()


print('Memory (After): {}MB'.format(mem_profile.memory_usage_resource())
print('Took{}Seconds'.format(t2-t1))
```
